```python
def dependency_parse(s):
    obj = { 'text': s }
    headers = {
        'content-type': 'application/json',
        'authorization': 'Bearer '+API_KEY
    }
    payload = json.dumps(obj)
    response = requests.request('POST', URL, data=payload, headers=headers)
    data = json.loads(response.text)
    return data
```
